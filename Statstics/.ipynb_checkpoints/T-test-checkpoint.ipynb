# T-test
A t-test is a type of inferential statistic which is used to determine if there is a significant difference between the means of two groups which may be related in certain features

- T-test has 2 types :
- one sampled t-test 
- two-sampled t-test 
- paired test







- DISTRIBUTION OF T-TEST

As sampling size increase t-distribution will get closer to standard sample distribution




# One sample T-test

```python
height = [10,20,35,50,28,40,55,18,16,55,30,25,43,18,30,28,14,24,16,17,32,35,26,27,65,18,43,23,21,20,19,70,45,78,56,48,23,72,77,92]
```

```python
len(height)
```

```python
import  numpy as np
```

```python
height_mean = np.mean(height)
print(height_mean) 
```

- it is mean of complete population 
- lets take some sample and find mean

```python
sample_size = 10
height_sample = np.random.choice(height,sample_size)
```

```python
height_sample
```

- here we will have Hypothesis Ho(Null) and Ha(alternate)

-  weather the diff btw sample mean and poulation mean

- Ho Null : no diff
- Ha alter : diff

if p-value < 0.05 then we will reject null hypo



```python
from scipy.stats import ttest_1samp
```

```python
ttest,p_value = ttest_1samp(height_sample,36)
```

36 is given to see weather the mean is 36 or not
- if it is 36 means null is true
- if not 36 means differnace

- 36 is taken from population mean

```python
print(p_value)
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

## 2nd Example

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
np.random.seed(6)
school_ages=stats.poisson.rvs(loc=18,mu=35,size=1500)
classA_ages=stats.poisson.rvs(loc=18,mu=30,size=60)
```

```python
classA_ages.mean()
```

```python
_,p_value=stats.ttest_1samp(a=classA_ages,popmean=school_ages.mean())
```

```python
p_value
```

```python
school_ages.mean()
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

# two sample t-test



```python
np.random.seed(12)
ClassB_ages=stats.poisson.rvs(loc=18,mu=33,size=60)
ClassB_ages.mean()
```

```python
from scipy.stats import ttest_ind
```

```python
_,p_value=stats.ttest_ind(a=classA_ages,b=ClassB_ages,equal_var=False)
```

```python
p_value
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

# Paired T-test

-when we want to check how differnt samples from same group

```python
weight1=[25,30,28,35,28,34,26,29,30,26,28,32,31,30,45]
weight2=weight1+stats.norm.rvs(scale=5,loc=-1.25,size=15)
```

- here in example we have taken some weight
- after some year weight are increase

```python

print(weight1)
```

```python
print(weight2)
```

```python
weight_df=pd.DataFrame({"weight_10":np.array(weight1),
                         "weight_20":np.array(weight2),
                       "weight_change":np.array(weight2)-np.array(weight1)})
```

```python
weight_df
```

```python
from scipy.stats import ttest_rel
```

```python
_,p_value = ttest_rel(a=weight1,b=weight2)
```

```python
print(p_value)
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```
