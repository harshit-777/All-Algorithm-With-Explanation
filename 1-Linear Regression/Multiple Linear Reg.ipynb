```python
data = pd.read_csv('statup.csv',index_col=0)
```

```python
data.head()

```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
data.State.value_counts()
```

```python
states = pd.get_dummies(X['State'],drop_first=True)
```

```python
states.head()
```

```python
X = X.drop('State',axis=1)
```

```python
X = pd.concat([X,states],axis=1)
```

```python
X.head()
```

- Y = m1(R & D) + m2(adm) + m3(Marketing) + m4(Florida) + m5(NewYork)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

# R**2 

- R**2 = 1 - sum of residual/sum of mean


- sum of residual = 1/n * sum(y-y_pred) * 2


- sum of mean  = 1/n * sum(y-y_mean) * 2


- Rsquare valuecome between 0 and 1 ,usually we get more then 0.70

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_pred)
```
