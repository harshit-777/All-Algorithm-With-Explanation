# Multi-Colliniearity

Multicollinearity refers to a situation in which two or more explanatory variables in a multiple regression model are highly linearly related.

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
import statsmodels.api as sn
```
