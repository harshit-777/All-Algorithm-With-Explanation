### strftime - passing formate specification

```python
st.strftime('%Y-%m-%d')
```

- Four digit year====%Y
- Two-digit year =====%y
- Two-digit month [01, 12] =====%m
- Two-digit day [01, 31]=====%d
- Hour (24-hour clock) [00, 23]====%H
- Hour (12-hour clock) [01, 12]===%I
- Two-digit minute [00, 59]===%M
- Second [00, 61] (seconds 60, 61 account for leap seconds)====%S
- Weekday as integer [0 (Sunday), 6]====%w

# Frequencies and Date Offsets

Frequencies in pandas are composed of a base frequency and a multiplier. Base fre
quencies are typically referred to by a string alias, like 'M' for monthly or 'H' for
hourly. For each base frequency, there is an object defined generally referred to as a
date offset.


- from pandas.tseries.offsets import Hour,Minute

- four_hour = Hour(4)

```python
pd.date_range('2000-01-01', '2000-01-03 23:59', freq='4h')
```

```python
pd.date_range('2000-01-01', periods=10, freq='1h30min')
```

# Time Zone Localization and Conversion

```python
rng = pd.date_range('3/9/2019 9:30', periods=6, freq='D')
```

```python
time_conv = pd.Series(np.random.randn(len(rng)),index=rng)
```

```python
time_conv
```

```python
time_conv_UTC = time_conv.tz_localize('UTC')
```

```python
time_conv_UTC
```

```python
time_conv_UTC.tz_convert('Asia/Kolkata')
```
