# Matplotlib
- is plotting library
- used is python
- numerical mathematics extension of numpy
- provide an object oriented api for plots
- cutomizable
- high quality


```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

if we are writing %matplotlib inline then we dont need to use plt.show()

```python
x = np.arange(0,10)
y = np.arange(20,30)
```

```python
a = np.arange(40,50)
b = np.arange(60,70)
```

## Scatter


```python
plt.scatter(x,y,c='r',marker='x')
plt.xlabel('X-points')
plt.ylabel('Y-points')
plt.title('Random Point Graph')
```

```python
# to save figure
plt.savefig('Ran.png')
```

## Plot
- will give simple stright line

```python
plt.plot(x,y,linestyle='--',linewidth=2,markersize=10)
```

## Subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
w = np.arange(0,6*np.pi,0.1)
k = np.sin(w)
plt.plot(w,k,'b--')
```

## Bar plot

```python
plt.bar(x,y,color='g')
```

## Hist

```python
l = np.array([12,58,76,54,53,70,100,54,44,48,57])
plt.hist(l)
```

## BOXplot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
```

```python
data
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
```

## Piechart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
