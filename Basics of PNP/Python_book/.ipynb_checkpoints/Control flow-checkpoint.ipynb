# If-else

```python
x = 4 
```

```python
if x < 0:
    print("It's negative")
elif x == 0:
    print('Equal to zero')
elif 0 < x < 5:
    print('Positive but smaller than 5')
else:
    print('Positive and larger than or equal to 5')
```

# For loop

## Continue

```python
s = [1,2,3,4,5,6,None,7]
t = 0

for i in s: 
    if i is None:
        continue
    t+=i  
    print(t)
```

## Break

```python
k = [1,2,3,4,5,6,None,7]
t = 0

for i in k: 
    if i is None:
        break
    t+=i  
    print(t)
```

# While

```python
x = 25
l =0 
while x > 0:
    if l > 50:
        break
    l+=x  
    print(l)
```

```python
sum = 0
for i in range(10):
# % is the modulo operator
    if i % 3 == 0 or i % 5 == 0:
        sum +=i
        print(sum)
```
