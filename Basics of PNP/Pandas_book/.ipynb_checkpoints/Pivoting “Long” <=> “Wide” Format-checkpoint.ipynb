# Long-Wide stacked

- A common way to store multiple time series in databases and CSV is in so-called long or stacked format

```python
df = pd.read_csv('long_stack.csv',index_col=0)
df
```

```python
periods = pd.PeriodIndex(year=df.year,quarter=df.quarter,name='date')
```

```python
periods
```

```python
col = pd.Index(['realgdp', 'infl', 'unemp'],name='item')
```

```python
df = df.reindex(columns=col)
```

```python
df
```

```python
df.index = periods.to_timestamp('D','end')
```

```python
df
```

```python
data = df.stack().reset_index().rename(columns={0:'value'})
```

```python
data.head(15)
```

```python
pivot = data.pivot('date','item','value')
```

```python
pivot
```

# Wide-Long Formate


```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
.....:
'A': [1, 2, 3],
.....:
'B': [4, 5, 6],
.....:
'C': [7, 8, 9]})
```

```python
df
```

```python
melted = pd.melt(df,['key'])
```

```python
melted
```

```python
reshaped = melted.pivot('key','variable','value')
```

```python
reshaped
```

```python
reshaped.reset_index()
```

```python
pd.melt(df, id_vars=['key'], value_vars=['A', 'B'])
```
