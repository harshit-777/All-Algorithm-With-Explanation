```python
dup_df
```

```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon','Pastrami', 'corned beef', 'Bacon','pastrami', 'honey ham', 'nova lox'],
                               'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
```

```python
data
```

```python
lower = data['food'].str.lower()
```

```python
lower
```

lets add one col to the food which eat by animal

```python
food_to_animal = {'bacon': 'pig',
                  'pulled pork': 'pig',
                  'pastrami': 'cow',
                  'corned beef': 'cow',
                  'honey ham': 'pig',
                  'nova lox': 'salmon'
                  }
```

# MAPPING

```python
data['animal'] = lower.map(food_to_animal)
```

```python
data
```

```python
# Alternate way
data['animal_lambda']=data.food.map(lambda x:food_to_animal[x.lower()])
```

```python
data
```

```python
data3  = pd.DataFrame(
                      np.arange(12).reshape((3,4)),
                      index=['aa','bb','cc'],
                      columns=['one','two','three','four']
                      )
```

```python
data3
```

```python
capital_word = lambda x:x[:4].upper()
```

```python
data3.index.map(capital_word)
```

```python
data3.index = data3.index.map(capital_word)
```

```python
data3
```

```python
data3 = data3.rename(columns=str.upper)
```
