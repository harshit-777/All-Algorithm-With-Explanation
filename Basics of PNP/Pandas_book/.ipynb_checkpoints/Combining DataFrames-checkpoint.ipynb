# Merge



```python
df1 = pd.DataFrame({'Name':['aayu','aayu','har','har'],
                    'Number':range(4)})
```

```python
df1
```

```python
df2 = pd.DataFrame({'Name':['AAYUSH','aayu','har'],
                   'Number1':range(3)})
```

```python
df2 
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='Name')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='left',on='Name')
```

```python
pd.merge(df1,df2,how='inner')
```

- When youre joining columns-on-columns, the indexes on the passed DataFrame objects are discarded.

# Example 2

```python
left = pd.DataFrame({'key1': ['Love', 'Love', 'Hate'],
                     'key2': ['Maths', 'Chemistry', 'Maths'],
                     'Student_A': [1, 2, 3]})
```

```python
right = pd.DataFrame({'key1': ['Love', 'Love', 'Hate', 'Hate'],
                      'key2': ['Maths', 'Maths', 'Chemistry', 'Chemistry'],
                      'Student_B': [4, 5, 6, 7]})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on=['key1','key2'],how='outer')
```

```python
pd.merge(left,right,on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

# Merging On Index

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})
```

```python
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
pd.merge(left1,right1,right_index=True,left_on='key',how='outer')
```

# Example 4

```python
lefth = pd.DataFrame({'Social_network': ['Insta', 'Insta', 'Insta','Whatsapp', 'Whatsapp'],

                      'download_year': [2010, 2011, 2012, 2011, 2012],

                       'data': np.arange(5.)})
```

```python
righth = pd.DataFrame(np.arange(12).reshape((6, 2)),
                      index=[['Whatsapp', 'Whatsapp', 'Insta', 'Insta','Insta', 'Insta'],
                                [2011, 2010, 2010, 2010, 2011, 2012]],
           columns=['Social_netwok1', 'Social_network2'])
```

```python
lefth
```

```python
righth
```

```python
pd.merge(lefth,righth,left_on=['Social_network','download_year'],right_index=True)
```

```python
pd.merge(lefth,righth,left_on=['Social_network','download_year'],right_index=True,how='outer')
```

```python
arr 
```

```python
np.concatenate([arr,arr],axis=1)
```

```python
s1= pd.Series([0,1],index=['a','b'])
s2= pd.Series([2,3,4],index=['c','d','e'])
s3= pd.Series([5,6],index=['f','g'])
```

```python
pd.concat([s1,s2,s3])
```

```python
pd.concat([s1,s2,s3],axis=1)
```

```python
r = pd.concat([s1,s2,s3],keys=['one','two','three'])
```

```python
r 
```

```python
r.unstack()
```

```python
re = pd.concat([s1,s2,s3],keys=['one','two','three'],axis=1)
```

```python
re
```

# Combining data with Overlap

```python
df3 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],
.....:
'b': [np.nan, 2., np.nan, 6.],
.....:
'c': range(2, 18, 4)})
```

```python
df4 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
.....:
'b': [np.nan, 3., 4., 6., 8.]})
```

```python
df3
```

```python
df4
```

```python
df3.combine_first(df4)
```
