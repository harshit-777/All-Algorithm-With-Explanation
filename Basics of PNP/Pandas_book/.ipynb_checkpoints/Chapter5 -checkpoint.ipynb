# Sorting and Ranking



```python
data4 = pd.Series(range(4), index=['d', 'a', 'b', 'c'])
```

```python
data4.sort_index()
```

```python
data4.sort_values()
```

```python
data5 = pd.Series([7, -5, 7, 4, 2, 0, 4])
```

```python
data5.rank()  #by default rank breaks ties by assigning each group the mean rank:
```

```python
data5.rank(method='first')
```

Here, instead of using the average rank 6.5 for the entries 0 and 2, they instead have
been set to 6 and 7 because label 0 precedes label 2 in the data.
