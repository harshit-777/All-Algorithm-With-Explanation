# Average

```python
df = pd.DataFrame({'category': ['a', 'a', 'a', 'a',
.....:
'b', 'b', 'b', 'b'],
.....:
'data': np.random.randn(8),
.....:
'weights': np.random.rand(8)})
```

```python
df 
```

```python
g = df.groupby('category')
```

```python
g 
```

```python
g_avg = lambda g: np.average(g['data'],weights=g['weights'])
```

```python
g.apply(g_avg)
```

# Correlation

```python
stock_df =  pd.read_csv('https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/examples/stock_px_2.csv',index_col=0,parse_dates=True)
```

```python
stock_df
```

```python
stock_df.info()
```

```python
stock_df.tail()
```

```python
spx_corr = lambda x: x.corrwith(x['SPX'])
```

```python
r = stock_df.pct_change().dropna()
```

```python
r.head() 
```

```python
get_year = lambda x: x.year
```

```python
by_year = r.groupby(get_year)
```

```python
by_year
```

```python
by_year.apply(spx_corr)
```

```python
by_year.apply(lambda g:g['AAPL'].corr(g['MSFT']))
```
