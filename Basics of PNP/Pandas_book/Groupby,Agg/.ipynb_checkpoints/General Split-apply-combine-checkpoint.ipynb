# General GroupBy 

- The most general-purpose GroupBy method is apply ,

```python
tips = pd.read_csv('group_tips.csv',index_col=0)
```

```python
tips.head()
```

```python
tips['tip_pct'] = tips['tip']/tips['total_bill']
```

```python
tips.head()
```

```python
def top(df,n=5,column='tip_pct'):
    return df.sort_values(by=column)[-n:]
```

```python
top(tips)
```

```python
tips.groupby('smoker').apply(top)
```

### apply on total_bill

```python
tips.groupby(['smoker','day']).apply(top,n=2,column='total_bill')
```

### Describe /

```python
result = tips.groupby('smoker')['tip_pct'].describe()
```

```python
result
```

```python
result.unstack()
```

#### Using lambda to describe

```python
result_lam = tips.groupby('smoker')['tip_pct'].apply(lambda x:x.describe())
```
