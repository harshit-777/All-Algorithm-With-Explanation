## Confusion Matrix

A confusion matrix is a table that is often used to describe the performance of a classification model (or classifier) on a set of test data for which the true values are known. It allows the visualization of the performance of an algorithm.

Aim -- Reducing type one and type 2 error
-  Sensitivity also called Recall --out of total actual postive value how many postivie we predicted correct


- Precision(positive prediction value) - out of toatal actual postive how many were actual postive 


- If more false postive are predicted then use precision and reduce the value


- If more False negative are  predicted then use Recall and reduce the value

## F-beta Score



How to choose beta

- beta = 1   if FN=TP are equally important
- beta = 0.5  if FP > FN (precision)
- beta = 2    if FP < FN (recall)

##  ROC and AUC

A receiver operating characteristic curve, or ROC curve, is a graphical plot that illustrates the diagnostic ability of a binary classifier system as its discrimination threshold is varied.



AUC is an abbrevation for area under the curve. It is used in classification analysis in order to determine which of the used models predicts the classes best.

```python
from sklearn import metrics
```

```python
C = 'Cat'
D = 'Dog'
F = 'Fox'
```

```python
y_true = [C,C,C,C,C,C, F,F,F,F,F,F,F, D,D,D,D,D]
```

```python
y_pred = [C,C,C,C,D,F ,F,F,F,D,D,C,F ,D,D,F,C,D]
```

```python
print(metrics.confusion_matrix(y_true,y_pred))
```
